```
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score as r2
import matplotlib.pyplot as plt
from numpy.fft import fft2 as fft
from numpy.fft import ifft2 as ifft
from tqdm import tqdm


plt.rcParams["font.size"] = 18
plt.rcParams['figure.figsize'] = (15,5)

from google.colab import drive
drive.mount("/content/drive")
import os 
os.chdir("/content/drive/My Drive/UT/laboratory/soft sensor")

df = pd.read_csv("SRU_data.txt", skiprows=1, header=None,sep="  ", dtype="float64",
                 names=["u1","u2","u3","u4", "u5", "y1", "y2"] )


X = df.iloc[:,:5].values
y1 = df.iloc[:,5].values
y2 = df.iloc[:,6].values

X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.4, shuffle=False)
X_valid, X_test, y1_valid, y1_test, y2_valid, y2_test = train_test_split(X_test, y1_test, y2_test, test_size=0.5, shuffle=False) 
```

```
# 時間領域プロット
plt.figure()
plt.plot(X_train)

# 周波数領域プロット
fourier_transfar = fft(X_train)
plt.figure()
for i in range(fourier_transfar.shape[1]):
    plt.plot(fourier_transfar.real[1:50,i]) 

# 周波数フィルタリング処理(周波数範囲:0～50)後の時間領域プロット
rfft = ifft(fourier_transfar[:100], s = (X_train.shape[0], X_train.shape[1])).real
plt.figure()
plt.plot(rfft)

for i in range(X_train.shape[1]):
    print(r2(X_train[:,i], rfft[:,i]))
```

```
# 周波数によるフィルタリング処理
r2_matrix = np.empty(shape=X_train.shape)
for i in range(X_train.shape[0]):   
    rfft = ifft(fourier_transfar[:i+1], s=X_train.shape).real 
    for j in range(X_train.shape[1]):
        r2_matrix[i,j] = r2(X_train[:,j], rfft[:,j])
```
